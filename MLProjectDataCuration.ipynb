```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from astropy.table import Table
```

```python
xxl_raw_data = pd.read_csv('agndata/agn_sample_observations.csv')
xxl_raw_data
```

```python
#Sort on ascending distance (off axis angle) and keep only rows with the lowest off axis angle
xxl = xxl_raw_data.sort_values('distance', ascending=True).drop_duplicates(subset='XXL_ID', keep='first').reset_index(drop=True, inplace=False)
xxl
```

```python
#OAA is still large for some sources, we'll place a generous cut at 10
xxl=xxl[xxl['distance']<10].reset_index(drop=True, inplace=False)
#We'll also cut on redshift z<3
xxl=xxl[xxl['z']<3].reset_index(drop=True, inplace=False)
print(len(xxl))
```

```python
#Save IDs and RA/Dec for SDSS query
xxl[['XXL_ID', 'RA_OPT', 'DEC_OPT']].to_csv('agndata/XXL_cutIDs.csv', index=False)
```

```python
#reading in manga agn data to see format
manga = Table.read('agndata/manga_agn-v2_0_1.fits')
#writing IDs for matching to photo objects
manga[['MANGAID', 'RA', 'DEC']].write('agndata/manga_agn_ids.csv', overwrite=True)
#remove trailing whitespace from IDs
manga['MANGAID'] = [manga['MANGAID'][i].strip() for i in range(len(manga))]
```

```python
manga
```

```python
# reading in and cleaning up positive train datasets
xxl_features = pd.read_csv('agndata/xxl_features_zreeves1.csv')
manga_features=pd.read_csv('agndata/manga_features_zreeves1.csv')

#adding column to designate AGN
xxl_features['is_AGN'] = [1]*len(xxl_features)

#adding column to designate AGN and how the AGN was detected
#this is stored in binary and converted to an integer
#0 is not AGN, 1 is radio, 2 is IR, 4 is optical, 8 is x-ray so 0100 -> 2 indicates an AGN detected in IR
#Manga AGN can have multiple detection methods, so 0101 -> 10 indicates an AGN detected in IR and X-ray
xxl_features['which_AGN'] = [8]*len(xxl_features) # 0 is not AGN, 1 is radio, 2 is IR, 4 is optical, 8 is x-ray

manga_features['is_AGN'] = [1]*len(manga_features)
which_AGN=[]
for i in range(len(manga_features)):
    manga_index = np.where(manga['MANGAID']==manga_features['MANGAID'][i])[0][0]
    sum_AGN = 1*manga['RADIO_AGN'][manga_index] + 2*manga['WISE_AGN'][manga_index] + 4*manga['BROAD_AGN'][manga_index] + 8*manga['BAT_AGN'][manga_index]
    which_AGN.append(sum_AGN)
manga_features['which_AGN'] = which_AGN

# setting catalog ID column to cleanly stack datasets
xxl_features['ext_ID'] = xxl_features['XXL_ID']
manga_features['ext_ID'] = manga_features['MANGAID']
```

```python
xxl_features.columns
```

```python
manga_features.columns
```

```python
#computing emission line flux ratios for xxl and manga catalogs

#XXL
xxl_features['o3_hB'] = [np.log10( xxl_features['Flux_OIII_5006'][i] / xxl_features['Flux_Hb_4861'][i]) for i in range(len(xxl_features))]
xxl_features['n2_hA'] = [np.log10( xxl_features['Flux_NII_6583'][i] / xxl_features['Flux_Ha_6562'][i]) for i in range(len(xxl_features))]
xxl_features['s2_hA'] = [np.log10( (xxl_features['Flux_SII_6716'][i]+xxl_features['Flux_SII_6730'][i]) / xxl_features['Flux_Ha_6562'][i]) 
                         for i in range(len(xxl_features))]
xxl_features['o1_hA'] = [np.log10( xxl_features['Flux_OI_6300'][i] / xxl_features['Flux_Ha_6562'][i]) for i in range(len(xxl_features))]
xxl_features['o3_o2'] = [np.log10( xxl_features['Flux_OIII_5006'][i] / xxl_features['Flux_OII_3726'][i]) for i in range(len(xxl_features))]
xxl_features['ne3_o2'] = [np.log10( xxl_features['Flux_NeIII_3868'][i] / xxl_features['Flux_OII_3726'][i]) for i in range(len(xxl_features))]

#Manga
manga_features['o3_hB'] = [np.log10( manga_features['oiii_5007_flux'][i] / manga_features['h_beta_flux'][i]) for i in range(len(manga_features))]
manga_features['n2_hA'] = [np.log10( manga_features['nii_6584_flux'][i] / manga_features['h_alpha_flux'][i]) for i in range(len(manga_features))]
manga_features['s2_hA'] = [np.log10( (manga_features['sii_6717_flux'][i]+manga_features['sii_6731_flux'][i]) / manga_features['h_alpha_flux'][i]) 
                         for i in range(len(manga_features))]
manga_features['o1_hA'] = [np.log10( manga_features['oi_6300_flux'][i] / manga_features['h_alpha_flux'][i]) for i in range(len(manga_features))]
manga_features['o3_o2'] = [np.log10( manga_features['oiii_5007_flux'][i] / manga_features['oii_3726_flux'][i]) for i in range(len(manga_features))]
manga_features['ne3_o2'] = [np.log10( manga_features['neiii_3869_flux'][i] / manga_features['oii_3726_flux'][i]) for i in range(len(manga_features))]
```

```python
# reading in and cleaning up negative data
neg_features = pd.read_csv('agndata/negative_features_zreeves1.csv')

#designate not AGN
neg_features['is_AGN'] = [0]*len(neg_features)
neg_features['which_AGN'] = [0]*len(neg_features)

neg_features['ext_ID'] = neg_features['sample_type']

#Computing emission line flux ratios
neg_features['o3_hB'] = [np.log10( neg_features['oiii_5007_flux'][i] / neg_features['h_beta_flux'][i]) for i in range(len(neg_features))]
neg_features['n2_hA'] = [np.log10( neg_features['nii_6584_flux'][i] / neg_features['h_alpha_flux'][i]) for i in range(len(neg_features))]
neg_features['s2_hA'] = [np.log10( (neg_features['sii_6717_flux'][i]+neg_features['sii_6731_flux'][i]) / neg_features['h_alpha_flux'][i]) 
                         for i in range(len(neg_features))]
neg_features['o1_hA'] = [np.log10( neg_features['oi_6300_flux'][i] / neg_features['h_alpha_flux'][i]) for i in range(len(neg_features))]
neg_features['o3_o2'] = [np.log10( neg_features['oiii_5007_flux'][i] / neg_features['oii_3726_flux'][i]) for i in range(len(neg_features))]
neg_features['ne3_o2'] = [np.log10( neg_features['neiii_3869_flux'][i] / neg_features['oii_3726_flux'][i]) for i in range(len(neg_features))]
```

```python
neg_features.columns
```

```python
#stacking data

#desginating the columns we want
columns = ['ext_ID', 'objID', 'specObjID', 'z', 
           'u_mag', 'g_mag', 'r_mag', 'i_mag', 'z_mag',  #SDSS color magnitudes
           'w1mpro', 'w2mpro', 'w3mpro', 'w4mpro',       #WISE color magnitudes
           'o3_hB', 'n2_hA', 's2_hA', 'o1_hA', 'o3_o2', 'ne3_o2',  #emission flux ratios
           'is_AGN', 'which_AGN']                        #AGN indicator

train_dataset = pd.concat([xxl_features[columns], manga_features[columns], neg_features[columns]], axis=0).reset_index(drop=True, inplace=False)
```

```python
#computing SDSS and WISE colors

#SDSS
train_dataset['u-g'] = [train_dataset['u_mag'][i] - train_dataset['g_mag'][i] for i in range(len(train_dataset))]
train_dataset['u-r'] = [train_dataset['u_mag'][i] - train_dataset['r_mag'][i] for i in range(len(train_dataset))]
train_dataset['u-i'] = [train_dataset['u_mag'][i] - train_dataset['i_mag'][i] for i in range(len(train_dataset))]
train_dataset['u-z'] = [train_dataset['u_mag'][i] - train_dataset['z_mag'][i] for i in range(len(train_dataset))]
train_dataset['g-r'] = [train_dataset['g_mag'][i] - train_dataset['r_mag'][i] for i in range(len(train_dataset))]
train_dataset['g-i'] = [train_dataset['g_mag'][i] - train_dataset['i_mag'][i] for i in range(len(train_dataset))]
train_dataset['g-z'] = [train_dataset['g_mag'][i] - train_dataset['z_mag'][i] for i in range(len(train_dataset))]
train_dataset['r-i'] = [train_dataset['r_mag'][i] - train_dataset['i_mag'][i] for i in range(len(train_dataset))]
train_dataset['r-z'] = [train_dataset['r_mag'][i] - train_dataset['z_mag'][i] for i in range(len(train_dataset))]
train_dataset['i-z'] = [train_dataset['i_mag'][i] - train_dataset['z_mag'][i] for i in range(len(train_dataset))]

#WISE
train_dataset['w1-w2'] = [train_dataset['w1mpro'][i] - train_dataset['w2mpro'][i] for i in range(len(train_dataset))]
train_dataset['w1-w3'] = [train_dataset['w1mpro'][i] - train_dataset['w3mpro'][i] for i in range(len(train_dataset))]
train_dataset['w1-w4'] = [train_dataset['w1mpro'][i] - train_dataset['w4mpro'][i] for i in range(len(train_dataset))]
train_dataset['w2-w3'] = [train_dataset['w2mpro'][i] - train_dataset['w3mpro'][i] for i in range(len(train_dataset))]
train_dataset['w2-w4'] = [train_dataset['w2mpro'][i] - train_dataset['w4mpro'][i] for i in range(len(train_dataset))]
train_dataset['w3-w4'] = [train_dataset['w3mpro'][i] - train_dataset['w4mpro'][i] for i in range(len(train_dataset))]
```
